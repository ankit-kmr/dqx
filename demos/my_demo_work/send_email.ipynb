```python
import smtplib
from email.mime.text import MIMEText

def send_completion_email(table_name, sender_email, recipient_email, success=True):
    # --- Configuration ---
    app_password = "tqea avon drdx pgmn" # Get this from Google Security settings
    smtp_server = "smtp.gmail.com"
    smtp_port = 587

    if success:
        subject = f"DQX Check Success for Table {table_name}"
        body = f"<h2 style='color:blue;'>DQX Check Completed Successfully</h2><p>Table: <b style='font-size:18px;'>{table_name}</b></p>"
    else:
        subject = f"DQX Check Failure for Table {table_name}"
        body = f"<h2 style='color:red;'>DQX Check Failed</h2><p>Table: <b style='font-size:18px;'>{table_name}</b></p>"

    msg = MIMEText(body, 'html')
    msg['Subject'] = subject
    msg['From'] = sender_email
    msg['To'] = recipient_email

    try:
        with smtplib.SMTP(smtp_server, smtp_port) as server:
            server.starttls()
            server.login(sender_email, app_password)
            server.sendmail(sender_email, recipient_email, msg.as_string())
        print("Email sent successfully!")
    except Exception as e:
        print(f"Error: {e}")
```
