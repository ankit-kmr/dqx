```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType, DateType

# Sample Customer DataFrame
customer_schema = StructType([
    StructField("customer_id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("email", StringType(), True),
    StructField("city", StringType(), True)
])

customer_data = [
    (1, "Alice Smith", "alice@example.com", "New York"),
    (2, "Bob Johnson", "bob@example.com", "San Francisco"),
    (3, "Carol Lee", "carol@example.com", "Chicago")
]

customer_df = spark.createDataFrame(customer_data, schema=customer_schema)

customer_df.write.format("delta").mode("overwrite").saveAsTable("dqx_sandbox.dqx_bronze.customer")
display(table('dqx_sandbox.dqx_bronze.customer'))
```

```python
# Sample Product DataFrame
product_schema = StructType([
    StructField("product_id", IntegerType(), True),
    StructField("product_name", StringType(), True),
    StructField("category", StringType(), True),
    StructField("price", DoubleType(), True)
])

product_data = [
    (101, "Laptop", "Electronics", 1200.00),
    (102, "Desk Chair", "Furniture", 150.00),
    (103, "Coffee Maker", "Appliances", 80.00)
]

product_df = spark.createDataFrame(product_data, schema=product_schema)

product_df.write.format("delta").mode("overwrite").saveAsTable("dqx_sandbox.dqx_bronze.product")
display(table('dqx_sandbox.dqx_bronze.product'))
```

```python
# Sample Order DataFrame
order_schema = StructType([
    StructField("order_id", IntegerType(), True),
    StructField("customer_id", IntegerType(), True),
    StructField("product_id", IntegerType(), True),
    StructField("order_date", DateType(), True),
    StructField("quantity", IntegerType(), True)
])

order_data = [
    (1001, 1, 101, "2026-02-20", 1),
    (1002, 2, 102, "2026-02-21", 2),
    (1003, 3, 103, "2026-02-22", 1)
]

order_df = spark.createDataFrame(order_data, schema=order_schema)

order_df.write.format("delta").mode("overwrite").saveAsTable("dqx_sandbox.dqx_bronze.order")
display(table('dqx_sandbox.dqx_bronze.order'))
```
